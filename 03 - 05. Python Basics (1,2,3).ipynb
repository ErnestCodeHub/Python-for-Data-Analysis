### If, elif ,else

```python
x = 5

if x < 0:
    print("Its negative")
elif x == 0:
    print("It is equal to 0")
elif 0 < x < 5:
    print("Positive but smaller than 5")
else:
    print("Positive and larger than or equal to 5")
```

### For Loops


* For Loops is used when no specific condition is needed. Such as loop all the variavles in the list * 0.5

```python
#Syntax

for value in collections:
    do something with value
```

```python
sequence = [1,2,None,4,None,5]

total = 0

for value in sequence:
    if value is None:
        continue
    total += value

total
```

```python
sequence = [1,2,0,4,6,5,2,1]

total_til_5 = 0

for value in sequence:
    if value ==5:
        break
    total_til_5 += value

total_til_5
```

```python
for i in range(5):
    for j in range(4):
        if j > i:
            break
        print((i,j))
```

### While loop

* A while loop specifies a condition and a block of code to be execuetd until the condition evalutes to False or the loop is explicitly ended with break


```python
x = 256

total = 0
while x > 0:
    if total > 500:
        break
    total += x
    x = x // 2

total
```

```python
print(256 // 2)
print(128//2)
print(64//2)
print(32//2)
print(16//2)
print(8//2)

print(f"\nThe final total is: {256+128+64+32+16+8}")
```

### pass

* Do nothing
* pass is the "no-operation" statement in Python. It is only required because Python uses whitespace to delimit blocks:

```python
x = -100

if x < 0:
    print("negative!")
elif x == 0:
    pass
else:
    print('positive!')
```
